```python
police = pd.read_csv('fatal-police-shootings-data.csv')
police['city'] = police['city'].str.replace(r'city', '') #removes anything with a lower-cased ' city', done to try to clean up data
police['city'] = police['city'].str.lower() #makes everything lower case, done to try to clean up data
police['city'] = police['city'].str.replace(r'\s', '')
police
```

```python
'changes the numbers in the dictionary into a list with the number inside, allows me to add the state the shooting took place in to the kill count'
def make_list_of_city_kill_state(kill_count,city_state):
    lst=[]
    for i in range(len(kill_count)):
        lst.append([city_state[i][0],kill_count[i],city_state[i][1]])
    return lst

city_kill_state = make_list_of_city_kill_state(lst_csk,lst_csk_dict)
print(len(city_kill_state))
city_kill_state
```

```python
def autocorrect(user_word, valid_words, diff_function, limit):
    if user_word in valid_words:
        return user_word
    a=min(valid_words,key=lambda word: diff_function(user_word,word,limit))
    if diff_function(user_word,a,limit)>limit:
            return user_word
    else:
        return a
```
